```python
# 3rd method (recommended)
# when I call something like this 
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]);
```

- Figure - The base canvas of all matplotlib plots. The overall thing you're plotting is a Figure, often shortened to fig.
- Axes - One Figure can have one or multiple Axes, for example, a Figure with multiple suplots could have 4 Axes (2 rows and 2 columns). Often shortened to ax.
- Axis - A particular dimension of an Axes, for example, the x-axis or y-axis

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
# %matplotlib inline # Not necessary in newer versions of Jupyter (e.g. 2022 onwards)
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot (Figure and Axes)
fig, ax = plt.subplots(figsize=(5,3))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("simple-plot.png")
```

```python
# make a plot from dictionary
nut_butter_price = {
    "Almond butter": 10,
    "Peanut butter": 8,
    "Cashew butter": 12,
}
fig, ax = plt.subplots(figsize=(8, 3))
ax.bar(x=nut_butter_price.keys(), height=nut_butter_price.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

#### Two options for subplots

```python
# Subplot option 1 (enumerate)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 5)
)

# PLot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(x=nut_butter_price.keys(), height=nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2 (position)
fig, ax = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(12, 5)
)
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(x=nut_butter_price.keys(), height=nut_butter_price.values())
ax[1, 1].hist(np.random.randn(1000));
```

```python
import pandas as pd
```

```python
np.random.seed(123456)
ts = pd.Series(np.random.randn(29), index=pd.date_range("11/1/2025", periods=29))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales['Price'] = car_sales['Price'].str.replace(r'[\$,\.]', '', regex=True)
car_sales['Price'] = car_sales['Price'].str[:-2]
```

```python
car_sales['Sale Date'] = pd.date_range('1,1,2020', periods=len(car_sales))
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
```

```python
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x='Sale Date', y='Total Sales')

car_sales['Price'] = car_sales['Price'].astype(int)
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter');
```

| Función                  | ¿Qué hace?                                                  | Ejemplo                                                    |
| ------------------------ | ----------------------------------------------------------- | ---------------------------------------------------------- |
| **`np.random.randn()`**  | Genera números aleatorios con distribución normal N(0,1).   | `np.random.randn(3)  # → [ 0.4, -1.2, 0.1 ]`               |
| **`np.random.random()`** | Genera números aleatorios uniformes entre 0 y 1.            | `np.random.random(3)  # → [0.82, 0.14, 0.66]`              |
| **`np.random.seed()`**   | Fija la semilla para que los aleatorios sean reproducibles. | `np.random.seed(42); np.random.random()  # → 0.3745401188` |

#### pyplot: to something quickly
#### matplotlib OO: more advanced

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(
    kind='scatter',
    x='age',
    y='chol',
    c='target',
    figsize=(10, 5)
);
```

```python
## OO meethod mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 5))
over_50.plot(
    kind='scatter',
    x='age',
    y='chol',
    c='target',
    ax=ax
);
# ax.set_xlim([45, 100]);
```

```python
over_50.target.value_counts()
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
);

# Customize the plot
ax.set(
    title='Heart Disease and Cholesterol Levels',
    xlabel='Age',
    ylabel='Cholesterol'
)

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(
    over_50['chol'].mean(),
    linestyle='--'
)
```
